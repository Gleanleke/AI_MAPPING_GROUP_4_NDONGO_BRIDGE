```python
import pandas as pd
import random
```

```python
dataset=pd.read_csv('Book1.csv')
i=0
while i < len(dataset):
    j = i + 1
    while j < len(dataset):
        if dataset.iloc[i, 0] == dataset.iloc[j, 0]:
            dataset.iloc[i, 1] = ((dataset.iloc[i, 1]) + (dataset.iloc[j, 1])) / 2
            dataset.iloc[i, 2] = ((dataset.iloc[i, 2]) + (dataset.iloc[j, 2])) / 2
            dataset.drop(dataset.index[j], inplace=True)  
        else:
            j += 1
    i += 1

```

```python
from haversine import haversine
i=0 
distance=[]
while i<len(dataset):
    j=1
    while j<len(dataset):
        culumn1=(dataset.iloc[i,1],dataset.iloc[i,2])
        culumn2=(dataset.iloc[j,1],dataset.iloc[j,2])
        distance.append(haversine(culumn1,culumn2))
        j+=1
    i+=1

search_map={}
i=0
while i<len(dataset):
    search_map[i] = [distance[i],dataset.iloc[i,1],dataset.iloc[i,2], dataset.iloc[i,0]]
    i=i+1
for i in search_map:
    print( search_map[i])

```
